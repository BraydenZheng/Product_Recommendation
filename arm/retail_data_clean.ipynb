Row to column algorithm
@Source: https://blog.csdn.net/qq_40395868/article/details/122197759

```python
tmp = ori.set_index(["Invoice", "StockCode"])["Description"]
```

```python
tmp = tmp[~tmp.index.duplicated(keep='first')]
```

```python
tmp 
```

```python
tmp = tmp.unstack()
```

```python
tmp = tmp.iloc[:, :-1]
```

```python
tmp
```

```python
tmp[:] = np.where(tmp.notnull(), tmp.columns, tmp)
```

```python
# tmp[:] = np.where(tmp.isnull(), '', tmp)
```

```python
tmp = tmp.apply(lambda row: pd.Series(sorted(row.tolist(), key=lambda x: np.isnan(x) if isinstance(x, float) else 0), index=row.index), axis=1)
```

```python
tmp
```

```python
tmp[:] = np.where(tmp.isnull(), '', tmp)
```

```python
tmp
```

```python
tmp = tmp[1:10]
```
