```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

Using same data but without normalization from NaiveBayes tab

```python
df = pd.read_csv('data/sales_dt.csv')
df.head(5)
```

```python
X_col = df.drop('rating',axis=1).columns
X_col
```

```python
y_col = 'rating'
```

```python
X = df.drop('rating',axis=1).values
X
```

```python
y = df['rating'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

Tree 1

```python
model = DecisionTreeClassifier(criterion='gini', random_state=0, max_depth=3)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
fig, ax = plt.subplots(figsize=(10, 10), dpi=300)
plot_tree(model, filled=True, ax=ax, feature_names=X_col, class_names=y_col)
plt.savefig('img/tree1.png')
plt.show()
```

```python

cm = confusion_matrix(y_test, y_pred, labels = model.classes_)

disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()

plt.savefig('img/tree1_cm.png')
plt.show()
```

Tree 2

```python
model = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=3)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
